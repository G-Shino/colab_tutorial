```
#CN method
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5, 5, 1000)
y = np.linspace(-5, 5, 1000)
xx, yy = np.meshgrid(x, y)
cc = xx + yy * 1j
val = (2 + cc)/(2-cc)
val = abs(val)
zz = val ** 2

fig, ax = plt.subplots(1, 1, figsize=(6, 6))
ax.patch.set_facecolor('coral') 
ax.contourf(xx, yy, zz, levels=[0, 1], cmap="tab20c")
ax.set_title("CN method", fontsize=12)
ax.set_xlabel("Z(R)")
ax.set_ylabel("Z(I)")
fig.show()
plt.savefig("CN.jpg")
```

```
#2nd order RK method
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5, 5, 5000)
y = np.linspace(-5, 5, 5000)

xx, yy = np.meshgrid(x, y)
cc = xx + yy * 1j
val = 1 + cc + cc ** 2 / 2
val = abs(val)
zz = val ** 2

fig, ax = plt.subplots(1, 1, figsize=(6, 6))
ax.patch.set_facecolor('coral') 
ax.contourf(xx, yy, zz, levels=[0, 1], cmap="tab20c")
ax.set_title("2nd order RK method")
ax.set_xlabel("Z(R)")
ax.set_ylabel("Z(I)")
fig.show()
plt.savefig("RK2.jpg")
```

```
#4th order RK method
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5, 5, 5000)
y = np.linspace(-5, 5, 5000)
xx, yy = np.meshgrid(x, y)
cc = xx + yy * 1j
val = 1 + cc + cc ** 2 / 2 + cc ** 3 / 6 + cc ** 4 / 24
val = abs(val)
zz = val ** 2

fig, ax = plt.subplots(1, 1, figsize=(6, 6))
ax.patch.set_facecolor('coral') 
ax.contourf(xx, yy, zz, levels=[0, 1], cmap="tab20c")
ax.set_title("4th order RK method")
ax.set_xlabel("Z(R)")
ax.set_ylabel("Z(I)")
fig.show()
plt.savefig("RK4.jpg")
```

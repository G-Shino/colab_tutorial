# クラスにおけるプロパティの設定

```
'''Dog クラス'''
class Dog():
     
    '''初期化メソッド'''
    def __init__(self):
        self.__name = "John" #非公開変数 __name
        
    '''クラスブロック内で非公開変数を使ったメソッドを作ります。'''
    def dogname(self):
        return self.__name
     
    '''nameプロパティのgetter'''
    @property
    def name(self): #プロパティ名は name です。
        return self.__name #__nameの値を返す。
     
    '''nameプロパティのsetter'''
    @name.setter
    def name(self, value): #getter と同じプロパティ名にします。
        self.__name = value #__nameの値を変更する
```

```
dog = Dog()
```

```
type(dog)
```

```
dog.name
```

```
dog.name = "sary"
```

```
dog.name
```

## 浅いコピー

```
from copy import copy

lst0 = [None for _ in range (3)]
lst1 = lst0
lst2 = copy(lst0)  # lst0.copy() と完全に同じことをしてます。

lst0[0] = 0
lst0[1] = 1
lst0[2] = 2

print(lst0)
print(lst1)
print(lst2)
```

```
from copy import copy

lst0 = [[None for _ in range (3)] for _ in range(3)]
lst1 = lst0
lst2 = copy(lst0)

lst0[0][0] = 0
lst0[0][1] = 1
lst0[0][2] = 2

print(lst0)
print(lst1)
print(lst2)
```

# pythonによるデータ分析入門の気になったところ

```
states = ["     Alabama ", "Georgia!", "georgia", "FlOrIda", "south   carolina##", "West virginia?"]
```

```
import re
#!,#,?を削除(""に置き換える)
def remove_puctuation(value):
  return re.sub("[!#?]", "", value)

#stripで空白削除 titleで単語の先頭を大文字に
#関数を引数にしている
clean_ops = [str.strip, remove_puctuation, str.title]

def clean_strings(strings, ops):
  result = list()
  for value in strings:
    for function in ops:
      value = function(value)
    result.append(value)
  return result
```

```
clean_strings(states, clean_ops)
```

```
words = ["apple", "bat", "bar", "atom", "book"]
by_letter = dict()
for word in words:
  letter = word[0]
  #第一引数に指定したキーが存在しない場合、新たな要素が追加される。
  #第一引数に指定したキーがすでに存在している場合は、第二引数にどんな値を指定しても元のオブジェクトのまま変更されない
  #また、このとき、第一引数に指定したキーに対応する値を返却する
  by_letter.setdefault(letter, []).append(word)
print(by_letter)
```

## 配列の結合

```
#基本的にはnp.concatenateとstackでおけ
#ただ二次元配列に対しては、vstackやhstackも便利

#concatenateは結合 
#stackは新たな軸を積み重ねるイメージ
import numpy as np
a1 = np.ones((3,), dtype="int64")
a2 = np.full((3,), 2)
a_con = np.concatenate([a1, a2], axis=0)
print("concatenate")
print(a_con)
print("*" * 10)

a1 = np.ones((2, 3), int)
a2 = np.full((2, 3), 2)
a_con = np.concatenate([a1, a2], axis=0)
print("concatenate")
print(a_con)
print("*" * 10)
a_con = np.concatenate([a1, a2], axis=1)
print(a_con)
print("*" * 10)
```

```
a1 = np.ones((3,), dtype="int64")
a2 = np.full((3,), 2)
# print(a1)
# print("*" * 10)
# print(a2)
# print("*" * 10)

a_sta = np.stack([a1, a2], axis=0)
print("stack 1D")
print(a_sta)
print("*" * 10)
a_sta = np.stack([a1, a2], axis=1)
print(a_sta)
print("*" * 10)

print("stack 2D")
a1 = np.ones((3, 4), dtype="int64")
a2 = np.full((3, 4), 2)
a_sta = np.stack([a1, a2], axis=0)
print(a_sta)
print(a_sta.shape)
a_sta = np.stack([a1, a2], axis=1)
print(a_sta)
print(a_sta.shape)
a_sta = np.stack([a1, a2], axis=2)
print(a_sta)
print(a_sta.shape)
```

```
#np.blockは直感的　配置したとおりに結合される
a1 = np.ones((2, 3), dtype="int64")
a2 = np.full((2, 3), 2)
a_blo = np.block([a1, a2])
print(a_blo)
print("*****************")
a_blo = np.block([[a1, a2], [a2, a1]])
print(a_blo)
a3 = np.full(6, 3)
print("*****************************")
print(np.block([[a1, a2], [a3]]))
```

```
#np.vstack()で縦に結合 
#二次元配列だとnp.concatenate(***, axis=0)と一緒
a1 = np.ones((2, 3), dtype="int64")
a2 = np.full((2, 3), 2)
a_vst = np.vstack([a1, a2])
print(a_vst)
print("++++++++++++++++++")
#一次元配列は二次元配列に拡張される。np.stack(*****, axis=0)と一緒
a1 = np.ones(3, dtype="int64")
a2 = np.full(3,2)
a_vst = np.vstack([a1, a2])
print(a_vst)
a1 = np.ones((2, 3), dtype="int64")
a2 = np.full(3, 2)
a_vst = np.vstack([a1, a2])
print("++++++++++++")
print(a_vst)
```

```
#np.vstack()で縦に結合 
#二次元配列だとnp.concatenate(***, axis=1)と一緒
a1 = np.ones((2, 3), dtype="int64")
a2 = np.full((2, 3), 2)
a_hst = np.hstack([a1, a2])
print(a_hst)
print("++++++++++++++++++")
#一次元配列は、bnp.concatenate(*****, axis=0)と一緒
a1 = np.ones(3, dtype="int64")
a2 = np.full(3, 2)
a_hst = np.hstack([a1, a2])
print(a_hst)
```

```
def create_counter():
    c = 0
    def count():
        nonlocal c  # １つ外側の変数 c を束縛する。
        print(locals())
        c = c + 1
        return c
    print(locals())
    return count
  
count = create_counter()
count()
count()
count()
```

```
def func():
    x = 3
    def value(v):
        nonlocal x
        x = v
    def add(y):
        return y+x
    x = 5
    return value, add

v, f = func()
print(f(4)) # 9 = 5 + 4 
v(10) # x の値が 5 → 10 に更新される
print(f(4)) # 14 = 4 + 10
```

```
def func():
    x = 3
    print("func name space : ", locals())
    def add3(y):
        print("add3 name space : ", locals())
        return y+x
    return add3

f = func()
print(f(4)) # 7
```

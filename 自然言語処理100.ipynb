## 05.n-gram
- 与えられたシーケンス（文字列やリストなど）からn-gramを作る関数を作成せよ．この関数を用い，"I am an NLPer"という文から単語bi-gram，文字bi-gramを得よ．
- 任意の文字列や文書を連続したn個の文字で分割するテキスト分割方法． 特に，nが1の場合をユニグラム（uni-gram），2の場合をバイグラム（bi-gram），3の場合をトライグラム（tri-gram）と呼ぶ．

```
import numpy as np
def n_gram(sequence, n):
  result = list()
  repeats = len(sequence) - n + 1
  for i in range(repeats):
    gram = sequence[i:i+n]
    result.append(gram)
  return result

sentence = "I am an NLPer"
ngram = 2
print(n_gram(sentence, ngram))

print(n_gram(sentence, 3))

lis = np.arange(5)
print(n_gram(lis, 2))
```

## 06.集合
"paraparaparadise"と"paragraph"に含まれる文字bi-gramの集合を，それぞれ, XとYとして求め，XとYの和集合，積集合，差集合を求めよ．さらに，'se'というbi-gramがXおよびYに含まれるかどうかを調べよ．

```
def n_gram(sequence, n):
  repeat_times = len(sequence) - n + 1
  result = list()
  for i in range(repeat_times):
    gram = sequence[i:i+n]
    result.append(gram)
  return result

sen1 = "paraparaparadise"
sen2 = "paragraph"

X = set(n_gram(sen1, 2))
Y = set(n_gram(sen2, 2))
print(X)
print(Y)
```

```
print(X | Y) #和集合　全体
print(set.union(X, Y))
```

```
print(set.intersection(X, Y))
X & Y #積集合　共通部分
```

```
print(set.difference(X, Y))
X - Y #Xが残る差集合
```

```
print(set.difference(Y, X))
Y - X #Yが残る差集合
```

## 07.テンプレートによる文生成
引数x, y, zを受け取り「x時のyはz」という文字列を返す関数を実装せよ．さらに，x=12, y="気温", z=22.4として，実行結果を確認せよ．

```
def sentence_maker(x, y, z):
  return str(x) + "時の" + str(y) + "は" + str(z)

x = 12
y = "気温"
z = 22.4
print(sentence_maker(x, y, z))
```

```
from string import Template

def hoge(x, y, z):
  return '{hour}時の{target}は{value}'.format(hour=x, target=y, value=z)

def format_string(x, y, z):
  #テンプレートを作成し、substituteで置換する
  s = Template('$hour時の$targetは$value')
  return s.substitute(hour=x, target=y, value=z)

print(hoge(x, y, z))
print(format_string(x, y, z))
  
```

## 08.暗号文
与えられた文字列の各文字を，以下の仕様で変換する関数cipherを実装せよ．

・英小文字ならば(219 - 文字コード)の文字に置換
・その他の文字はそのまま出力

この関数を用い，英語のメッセージを暗号化・復号化せよ．

```
def cipher(sentence):
  new_sentence = str()
  for i in sentence:
    if i.islower():
      new_sentence += chr(219 - ord(i))
    else:
      new_sentence += i
  return new_sentence

cipher("I Am An Engineer")
```

## 09.Typoglycemia
スペースで区切られた単語列に対して，各単語の先頭と末尾の文字は残し，それ以外の文字の順序をランダムに並び替えるプログラムを作成せよ．ただし，長さが４以下の単語は並び替えないこととする．適当な英語の文（例えば"I couldn't believe that I could actually understand what I was reading : the phenomenal power of the human mind ."）を与え，その実行結果を確認せよ．

```
import random

def rnd_sentence(sentence):
  list_words = sentence.split(" ")
  new_sentence = list()
  for word in list_words:
    if len(word) <= 4:
      new_sentence.append(word)
      continue
    head = word[0]
    tail = word[-1]
    center = word[1:-1]
    #random.shuffleはimmutableには使用できない
    rnd_sen = "".join(random.sample(center, len(center)))
    new_sentence.append(head + rnd_sen + tail)
  return " ".join(new_sentence)


sentence = "I couldn't believe that I could actually understand what I was reading : the phenomenal power of the human mind ."
rnd_sentence(sentence)
```

## 12.1列目をcol1.txtに，2列目をcol2.txtに保存

```
import re
import numpy as np

list_lines = list()
with open("hightemp.txt", "r") as f:
  for line in f:
    line = re.sub("\n", " ", line)
    list_line_split = re.split("\t", line)
    list_lines.append(list_line_split)
  
list_lines = np.array(list_lines)
col1 = list_lines[:, 0]
col2 = list_lines[:, 1]
with open("col1.txt", "w") as f:
  col1 = "\n".join(col1)
  f.write(col1 + "\n")
with open("col2.txt", "w") as f:
  col2 = "\n".join(col2)
  f.write(col2 + "\n")
```

## 13.col1.txtとcol2.txtをマージ
12で作ったcol1.txtとcol2.txtを結合し，元のファイルの1列目と2列目をタブ区切りで並べたテキストファイルを作成せよ．確認にはpasteコマンドを用いよ．

```
import re
with open("col1.txt", "r") as f:
  list_col1_lines = f.readlines()
with open("col2.txt", "r") as f:
  list_col2_lines = f.readlines()

list_cols = list()
for word_col1, word_col2 in zip(list_col1_lines, list_col2_lines):
  word_col1 = re.sub("\n", "", word_col1)
  line_new = word_col1 + "\t" + word_col2
  list_cols.append(line_new)
result = "".join(list_cols)
with open("col1+col2.txt", "w") as f:
  f.write(result)
```

```
import numpy as np

with open("hightemp.txt", "r") as f:
  list_lines = list()
  for line in f:
    line = line.rstrip("\n")
    list_line = line.split("\t")
    list_lines.append(list_line)
list_lines = np.array(list_lines)
list_args_sorted = np.argsort(list_lines[:, 2])
list_lines_sorted = list_lines[list_args_sorted[::-1], :]
# list_lines_sorted = list_lines_sorted.flatten()
sen = "".join(["\t".join(line) + "\n" for line in list_lines_sorted])
print(sen.rstrip("\n"))
```

## 19.各行の1コラム目の文字列の出現頻度を求め，出現頻度の高い順に並べる
各行の1列目の文字列の出現頻度を求め，その高い順に並べて表示せよ．確認にはcut, uniq, sortコマンドを用いよ．

```
with open("hightemp.txt", "r") as f:
  list_lines = f.readlines()
  
from collections import defaultdict
dic_word = defaultdict(int)
for line in list_lines:
  word0 = line.split("\t")[0]
  dic_word[word0] += 1
dic_sorted = sorted(dic_word.items(), key=lambda x: x[1], reverse=True)
print(dic_sorted)
```
